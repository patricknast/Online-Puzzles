```python
def adjacent_rock_faces(coord,rock_coords):
    i,j,k = coord
    adjacent_coords = get_next_coords(coord)
    intersection = [c for c in adjacent_coords if c in rock_coords]
    obscured_faces = len(intersection)
    unobscured_faces = 6 - obscured_faces
    return obscured_faces, unobscured_faces
```

```python
def surface_area(search_coords,rock_coords):
    sa = 0
    for coord in tqdm(search_coords):
        num_obs, num_non_obs = adjacent_rock_faces(coord,rock_coords)
        sa += num_non_obs
    return sa

def exterior_surface_area(search_coords,rock_coords):
    sa = 0
    for coord in tqdm(search_coords):
        num_obs, num_non_obs = adjacent_rock_faces(coord,rock_coords)
        sa += num_obs
    return sa
```

```python
def full_grid(cube_size):
    full_grid = []
    for i in range(max_x +1):
        for j in range(max_y+1):
            for k in range(max_z + 1):
                full_grid.append((i,j,k))
    return full_grid
            
def inside_cube(coord,cube_size):
    i,j,k = coord
    return -1 <= i <= cube_size and -1 <= j <= cube_size and -1 <= k <= cube_size

def get_next_coords(coord):
    i,j,k = coord
    return [(i+1,j,k),(i-1,j,k),(i,j+1,k),(i,j-1,k),(i,j,k+1),(i,j,k-1)]
```

```python
class Rock:
    def __init__(self,rock_coords):
        self.rock_coords = rock_coords
        self.cube_size = max(coord[i] for coord in rock_coords for i in range(len(coord)))+ 1
        self.exterior_air_coords = [(0,0,0)]
    def find_exterior_air_coords(self,coord=(0,0,0)):
        next_coords = [next_coord for next_coord in get_next_coords(coord) if inside_cube(next_coord,self.cube_size) and not next_coord in self.rock_coords and not next_coord in self.exterior_air_coords]
        self.exterior_air_coords.extend(next_coords)
        for next_coord in next_coords:
            self.find_exterior_air_coords(next_coord)
    def full_surface_area(self):
        return surface_area(self.rock_coords,self.rock_coords)
    def exterior_surface_area(self):
        return exterior_surface_area(self.exterior_air_coords,self.rock_coords)
        
        
```
