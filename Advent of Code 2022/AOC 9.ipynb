```python
def move_tail(head_pos,tail_pos):
    diff = list(np.subtract(head_pos, tail_pos))
    #RIGHT
    if diff == [0,2]:
        tail_pos[1] += 1
    #LEFT
    elif diff == [0,-2]:
        tail_pos[1] -= 1
    #UP
    elif diff == [2,0]:
        tail_pos[0] += 1
    #DOWN
    elif diff == [-2,0]:
        tail_pos[0] -= 1
    #DIAG UL
    elif diff == [1,2] or diff == [2,1] or diff == [2,2]:
        tail_pos[0] += 1
        tail_pos[1] += 1
    elif diff == [-1,2] or diff == [-2,1] or diff == [-2,2]:
        tail_pos[0] -= 1
        tail_pos[1] += 1
    elif diff == [-1,-2] or diff == [-2,-1] or diff == [-2,-2]:
        tail_pos[0] -= 1
        tail_pos[1] -= 1
    elif diff == [1,-2] or diff == [2,-1] or diff == [2,-2]:
        tail_pos[0] += 1
        tail_pos[1] -= 1

    return tail_pos
def move_head(head_pos,command):
    if command[0] == 'U':
        head_pos[0] -= 1
    elif command[0] == 'D':
        head_pos[0] += 1
    elif command[0] == 'L':
        head_pos[1] -= 1
    elif command[0] == 'R':
        head_pos[1] += 1
    return head_pos
```

```python
from copy import deepcopy
class Knot:
    def __init__(self,position):
        self.position = position
        self.head_knot = None
        self.tail_knot = None

    def follow_head(self):
        if self.head_knot:
            self.position = move_tail(self.head_knot.position,self.position)
        if self.tail_knot:
            self.tail_knot.follow_head()
    def do_command(self,single_command):
        self.position = move_head_one(self.position, single_command)
        if self.tail_knot:
            self.tail_knot.follow_head()
    def create_tail(self):
        self.tail_knot = Knot(deepcopy(self.position))
        self.tail_knot.head_knot = self

def move_head_one(head_pos,single_command):
    if single_command == 'U':
        head_pos[0] -= 1
    elif single_command == 'D':
        head_pos[0] += 1
    elif single_command == 'L':
        head_pos[1] -= 1
    elif single_command == 'R':
        head_pos[1] += 1
    return head_pos        
```
