```python
views_list = []
num_rows = 99
for i in np.arange(0,num_rows):
    for j in np.arange(0,num_rows):
        current_index = array_list[i][j]
        view_score = 1
        #LOOK LEFT
        views_a = [p for p in np.arange(0,j) if array_list[i][p] >= current_index]
        if views_a:
            view_a =  j - max(views_a)
        else: view_a = j
        #LOOK RIGHT
        views_b = [p for p in np.arange(0,(num_rows-1)-j) if np.flip(array_list,1)[i][p] >= current_index]
        if views_b:
            view_b = ((num_rows-1)-j) - max(views_b)
        else: view_b = (num_rows-1)-j
        # LOOK UP
        views_c = [p for p in np.arange(0,i) if array_list.T[j][p] >= current_index]
        if views_c:
            view_c = i - max(views_c)
        else: view_c = i
        #LOOK DOWN
        views_d = [p for p in np.arange(0,(num_rows-1)-i) if np.flip(array_list.T,1)[j][p] >= current_index]
        if views_d:
            view_d = ((num_rows-1)-i) - max(views_d)
        else: view_d = (num_rows-1)-i
        view_score *= view_c*view_a*view_b*view_d
        #print((i,j),current_index,(views_c,views_a,views_b,views_d),(view_c,view_a,view_b,view_d))
        views_list.append(view_score)
print(max(views_list))
```
