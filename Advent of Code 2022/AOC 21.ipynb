```python
import numpy as np
from tqdm import tqdm
```

```python
monkey_dict = {}
for line in puzz_input:
    monkey_dict[line[0]] = line[1]
monkey_dict
```

```python
def parse_string(string):
    if string.isnumeric():
        return int(string)
    else:
        s1, op, s2 = string.split(' ')
    if op == '=':
        return parse_string(monkey_dict[s1]) == parse_string(monkey_dict[s2])
    if op == '+':
        return parse_string(monkey_dict[s1]) + parse_string(monkey_dict[s2])
    elif op == '-':
        return parse_string(monkey_dict[s1]) - parse_string(monkey_dict[s2])
    elif op == "*":
        return parse_string(monkey_dict[s1]) * parse_string(monkey_dict[s2])
    elif op == '/':
        return parse_string(monkey_dict[s1]) / parse_string(monkey_dict[s2])

    
```
