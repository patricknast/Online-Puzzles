```python
def type_check(val1,val2):
    type1,type2 = type(val1), type(val2)
    if type1 == type2 == list:
        return compare_lists(val1,val2)
    elif type1 == type2 == int:
        return compare_ints(val1,val2)
    elif type1 == list and type2 == int:
        return compare_lists(val1,[val2])
    elif type2 == list and type1 == int:
        return compare_lists([val1],val2)

def compare_ints(int1,int2):
    if int1 > int2:
        return False
    elif int2 > int1:
        return True
def compare_lists(list1,list2):
    try:
        for i in range(len(list1)):
            test_val = type_check(list1[i],list2[i])
            if test_val is not None:
                return test_val
        if len(list2) > len(list1):
            return True
    except IndexError:
        return False
```

```python
import numpy as np
from functools import cmp_to_key as ctk
with open('aoc13.txt') as f:
    lines = f.readlines()
lines = [line.strip() for line in lines]
lines = [eval(line) for line in lines if line]
```

```python
def type_check(val1,val2):
    type1,type2 = type(val1), type(val2)
    if type1 == type2 == list:
        return compare_lists(val1,val2)
    elif type1 == type2 == int:
        return compare_ints(val1,val2)
    elif type1 == list and type2 == int:
        return compare_lists(val1,[val2])
    elif type2 == list and type1 == int:
        return compare_lists([val1],val2)

def compare_ints(int1,int2):
    if int1 > int2:
        return 1
    elif int2 > int1:
        return -1
def compare_lists(list1,list2):
    try:
        for i in range(len(list1)):
            test_val = type_check(list1[i],list2[i])
            if test_val is not None:
                return test_val
        if len(list2) > len(list1):
            return -1
    except IndexError:
        return 1
```

```python
right_order = sorted(lines,key=ctk(compare_lists))
print((right_order.index([[6]])+1)*(right_order.index([[2]])+1))
```
